# Import libraries

```python
import pandas as pd
import sys
import os
%matplotlib inline
```

# Load Dataset

```python
def read_dataset(file):
    df = pd.read_csv(file)
    return df

def get_data_info(df):
    bengin = df['verdict'].value_counts()[0]
    malicious = df['verdict'].value_counts()[1]
    total_record = df.shape[0]
    malicious_percentage = round((malicious/total_record)*100)
    benign_percentage = round((bengin/total_record)*100)
    print("Total Dataset Records--", total_record)
    print("Total Features---------", df.shape[1])
    print("Total Benign Urls------", malicious)
    print("Total Malicious Urls---", bengin)
    print("Malicious Percentage---", malicious_percentage)
    print("Benign Percentage------", benign_percentage)
    return True
```

# Data Exploration

```python
def get_mean_values_for_all_features(df):
    print("*** Mean values of different attributes *** ")
    return df.groupby('verdict').mean()
#     return True

def plot_entropy_effect_on_class(df):
    df.groupby('verdict')['domain_entropy'].mean().plot(kind='bar')
    return True

def plot_is_html_effect_on_class(df):
    df.groupby('verdict')['url_is_html'].mean().plot(kind='bar')
    return True

def plot_domain_n_url_favicon_comparision_graph(df):
    print("Comparison of domain total favicons & Url total favicon")
    df.groupby('verdict')['domain_total_favicon','url_total_favicon'].mean().plot(kind='bar')
    return True

def plot_total_og_domains_effect_on_class(df):
    print("Comparison of domain total favicons & Url total favicon")
    df.groupby('verdict')['domain_total_og_domains'].mean().plot(kind='bar')
    return True

def plot_entropy_effect_on_class(df):
    df.groupby('verdict')['domain_entropy','url_entropy'].mean().plot(kind='bar')
    return True

def plot_total_og_links_effect_on_class(df):
    df.groupby('verdict')['domain_total_og_links'].mean().plot(kind='bar')
    return True

def plot_favicon_match_effect_on_class(df):
    df.groupby('verdict')['is_favicon_match'].mean().plot(kind='bar')
    return True

def effect_of_title_n_favicon_match_on_class(df):
    df.groupby('verdict')['title_match', 'is_favicon_match'].mean().plot(kind='bar')
    return True

def effect_of_uri_length_on_class(df):
    df.groupby('verdict')['uri_length'].mean().plot(kind='bar')
    return True

def get_attribute_effect_on_class(df, attribute_list=[]):
    df.groupby('verdict')[attribute_list].mean().plot(kind='bar')
    return True
```

# Main Function

```python
dataset_file = "/home/shahid/projects/behaviouralClassifier/classifier/urls_data_v1.csv"
df = read_dataset(dataset_file)
get_data_info(df)
get_mean_values_for_all_features(df)

# attribute plots
get_attribute_effect_on_class(df, ['domain_total_favicon','url_total_favicon'])
get_attribute_effect_on_class(df, ['domain_total_og_domains'])
get_attribute_effect_on_class(df, ['domain_entropy','url_entropy'])
get_attribute_effect_on_class(df, ['url_is_html','domain_is_html'])
get_attribute_effect_on_class(df, ['domain_total_og_links','url_total_og_links'])
get_attribute_effect_on_class(df, ['is_favicon_match'])
get_attribute_effect_on_class(df, ['uri_length', 'url_length'])
get_attribute_effect_on_class(df, ['title_match', 'is_favicon_match'])
```
